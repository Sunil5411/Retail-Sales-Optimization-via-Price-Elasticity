```python
import pandas as pd

# Load your CSV file
df = pd.read_csv('India_Retail_Elasticity_Dataset_1000Rows.csv')

# 1. View basic info
print(df.info())

# 2. Check for missing values
print(df.isnull().sum())

# 3. Drop duplicates if any
df = df.drop_duplicates()

# 4. Handle missing values (example: drop rows with missing critical data)
df = df.dropna(subset=['Product_ID', 'Final_Price', 'Units_Sold'])

# 5. Convert data types if needed
df['Date'] = pd.to_datetime(df['Date'])

# 6. Quick stats to understand data ranges
print(df[['Final_Price', 'Units_Sold', 'Revenue']].describe())

# 7. Optional: Remove or flag outliers in price or units sold
# Example: Filter out unreasonable prices (e.g., <0 or extremely high)
df = df[(df['Final_Price'] > 0) & (df['Final_Price'] < 10000)]

# Example: Filter units sold to a reasonable range
df = df[(df['Units_Sold'] > 0) & (df['Units_Sold'] < 1000)]

print(f"Cleaned dataset shape: {df.shape}")

# Save cleaned data for further analysis
df.to_csv('cleaned_retail_data.csv', index=False)
```
